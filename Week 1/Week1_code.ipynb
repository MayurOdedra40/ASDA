```python
# Split the string by spaces
words = input_string.split()

#create a list of alphabets
lowercaseAlphabets = list(string.ascii_lowercase)
# Create a dictionary to count each word
wordCount = {}

for word in words:
    wordLower = word.lower()  
    wordCount[wordLower] = wordCount.get(wordLower, 0) + 1    
```

```python
# Initialize an empty list to store dictionaries
wordList = []

for word in words:
    wordLower = word.lower() 
    # Create a dictionary for each word
    newWord = {
        'word': word,
        'length': len(word),
        'lowercase': wordLower,
        'OnlyWord': (word in lowercaseAlphabets), 
        'count': wordCount[wordLower]  
    }
    # add the dictionary to the list
    wordList.append(newWord)
```

```python

# Display the results
for info in wordList:
    print(info)
```
